```python
# Handling the missing values
df[['director','cast','country']] = df[['director','cast','country']].fillna('Unknown')
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df.dropna(axis=0, inplace = True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Convert all text columns to lowercase and strip whitespace
text_columns = df.select_dtypes(include='object').columns
df[text_columns] = df[text_columns].apply(lambda x: x.str.lower().str.strip())
```

```python
#Standardize country names
country_replacements = {
    'united states': 'usa',
    'united kingdom': 'uk',
    'south korea': 'korea',
    'russia': 'russian federation',
    'u.s.': 'usa',
    'us': 'usa',
    'england': 'uk',
    'india': 'india',
    }
df['country'] = df['country'].replace(country_replacements)
```

```python
#Standardize gender values
if 'gender' in df.columns:
    gender_replacements = {
        'male': 'male',
        'm': 'male',
        'female': 'female',
        'f': 'female',
        'non-binary': 'nonbinary',
        'nb': 'nonbinary',
    }  
    df['gender'] = df['gender'].replace(gender_replacements)
```

```python
df = df.drop_duplicates()
```

```python
df['date_added'] = df['date_added'].astype(str).str.strip().str.title()

# Convert to datetime64
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')

print("Actual dtype:", df['date_added'].dtype)  # Should show datetime64[ns]

print("date added")
print(df['date_added'].dt.strftime('%d-%m-%Y').head())
```

```python
# Clean column headers: lowercase, replace spaces with underscores, remove special characters
df.columns = (
    df.columns
    .str.lower()                      # Convert to lowercase
    .str.strip()                      # Remove leading/trailing spaces
    .str.replace(' ', '_')            # Replace spaces with underscores
    .str.replace(r'[^\w\s]', '', regex=True)  # Remove special characters
)

print(df.columns.tolist())
```
